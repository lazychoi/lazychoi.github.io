```python
arr = [2, 8, 6, 1, 10, 15, 3, 12, 11]
# arr = [2, 8, 6, 1, 10]

# 권오흠 교수 강의 듣고 구현하다가 실패한 코드
# def max_heapify(arr, parent):
#     print(f'heapify {parent} => {arr}')
#     larger = parent
#     left = 2*parent+1           # 왼쪽 자식 노드
#     right = 2*parent+2          # 오른쪽 자식 노드
#     # if left >= len(arr)-1  and right >= len(arr):  # 자식 노드가 없으면 재귀 호출 멈춤
#     #     print("자식 노드가 없으므로 멈춤 =================")
#     #     return
#     if right < len(arr) and arr[left] >= arr[right]:    # 자식 노드가 존재하고 왼쪽 노드가 오른쪽 노드보다 크면 왼쪽 노드 인덱스를 larger에 저장
#         larger = left
#         print("왼쪽 노드 인덱스", larger)
#     elif right < len(arr) and arr[right] > arr[parent]:         # 오른쪽 노드가 존재하고 부모보다 크면 오른쪽 노드 인덱스를 larger에 저장
#         larger = right
#         print("오른쪽 노드 인덱스", larger)
#     if arr[parent] < arr[larger]:          # 부모 노드가 자식 노드보다 크면 재귀 멈춤
#         arr[parent], arr[larger] = arr[larger], arr[parent]  # 부모 노드와 큰 자식 노드를 교환
#         return
#     print(f'부모 노드({arr[larger]})와 자식 노드({arr[parent]}) 교환 => {arr}')
#     max_heapify(arr, larger)             # 교환한 자식 노드에서 재귀 호출

# def max_heapify(arr, n, i):
#     k = i            # 부모 노드를 큰 값을 저장하는 k변수에 저장
#     while(2*i+1 < n and 2*i+2 < n):
#         left = 2*i+1     # left child
#         right = 2*i+2    # right child
#         print(f"왼쪽 자식={left}({arr[left]}), 오른쪽 자식={right}({arr[right]})")
#         if arr[left] > arr[right]: k = left       # 왼쪽 자식이 크면 그 인덱스를 k에 저장
#         elif arr[right] > arr[left]: k = right    # 오른쪽 자식이 크면 
#         if arr[i] >= arr[k]: return               # 부모가 크면 다음 노드로 넘어가거나 중지
#         arr[i], arr[k] = arr[k], arr[i]
#         i = k
#         print("교환후 배열=>", arr)
#         print("\ni=", i, "n=", n, "right", 2*i+2)


def max_heapify(arr, n, i):
    '''
    마지막 leaf node의 부모(n/2-1)부터 역순으로 노드를 검사하여 heap을 만듦
    * 주어진 배열을 max heap로 만드는 함수
    * arr : 배열
    * n : max heap으로 만들 범위 -> heap sort에서 root를 맨 끝으로 보낸 후 제외하기 위해 필요
    * i : 비교할 부모 노드 인덱스
    '''
    
    large = i           # 부모 노드가 가장 크다고 가정
    left = 2*i+1        # 왼쪽 자식 노드
    right = 2*i+2       # 오른쪽 자식 노드
    if left < n and arr[left] > arr[large]:     # 왼쪽 자식이 검사할 배열(n) 내에 있고 부모 노드보다 크면
        large = left
        print(f"왼쪽 자식 index={left}({arr[left]})")
    if right < n and arr[right] > arr[large]:   # 오른쪽 자식이 검사할 배열(n) 내에 있고 기존의 가장 큰 노드보다 크면(형제끼리 비교 가능)
        large = right
        print(f"오른쪽 자식 index={right}({arr[right]})")
    if large != i:                              # 가장 큰 값이 부모와 다르면 교환
        arr[i], arr[large] = arr[large], arr[i]
        print("교환후 배열=>", arr)
        print(f"{large}({arr[large]})를 새로운 부모로 하여 재귀 호출")
        max_heapify(arr, n, large)


def heap_sort(arr: list):
    n = len(arr)

    # maxheap 만들기
    for i in range( n//2 -1, -1, -1):  # 마지막 노드의 부모(n/2 -1)부터 시작하여 거꾸로 돌기
        print("====== max heap 만들기 =======")
        print(f"부모 노드 index={i}({arr[i]}) => {arr}")
        max_heapify(arr, n, i)

    print("***** max heap 생성 완료 *****")
    
    # 첫번째 요소를 맨 뒤로 보내며 힙 만들기
    for i in range(n-1, 0, -1):  # 힙 크기에서 시작해 1까지
        print("\n=========== 정렬 ===============")
        print(arr)
        arr[0], arr[i] = arr[i], arr[0]  # 배열의 첫(root node)과 끝(end node)을 바꾼다
        print(f"맨 앞과 맨 뒤({i}) 교환(가장 큰 수를 맨 뒤로 보낸) 후\n=>", arr)
        print("+++++ 힙 만들기 +++++")
        print(f"마지막 index={i}(=> 함수에서 인덱스 {i-1}까지 heapify)")
        max_heapify(arr, i, 0)      # 마지막 요소를 제외하려면 마지막 인덱스를 가리키는 i값을 매개변수에 넣어야 함.

heap_sort(arr)
print("\n최종 배열=> ", arr)
```
