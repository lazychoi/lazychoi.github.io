```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(url, index_col=0)
df.head(3)
```

```python
df['newIncome'] = df['Income_Category'].map({'Unknown' : 'N', 
                                             'Less than $40K' : 'a', 
                                             '$40K - $60K' : 'b', 
                                             '$60K - $80K' : 'c',
                                             '$80K - $120K' : 'd',
                                             '$120K +’' : 'e'})
df.head(3)
```

```python
# 위와 같은 코드
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}

df['newIncome']  =df.Income_Category.map(lambda x: dic[x])
df['newIncome'].head(3)
```

```python
def age(n):
    if n >=70: return 7
    elif n >= 60: return 6
    elif n >= 50: return 5
    elif n >= 40: return 4
    elif n >= 30: return 3
    elif n >= 20: return 2
    elif n >= 10: return 1
    else: return 0
df['AgeState'] = df['Customer_Age'].apply(age)
df['AgeState'].value_counts().sort_index()
```

Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라

```python
# apply 대신 map을 써도 됨
df['newEduLevel'] = df['Education_Level'].apply(lambda x: 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()
```

Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라

```python
df['newLimit'] = df['Credit_Limit'].map(lambda x: 1 if x >= 4500 else 0)
df['newLimit'].value_counts()
```

컬럼 두 개에 apply 적용. axis=1 매개변수를 넣어야 한다는 것을 주의한다.

Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

주의!! (x.year - 100)을 하면 숫자로 바뀐다. 따라서 숫자를 datetime 형식으로 바꾼다.

```python
import datetime

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(lambda x: 
                                      pd.to_datetime(datetime.date(x.year - 100, x.month, x.day)) 
                                      if x.year >= 2061 else x)
df.head(3)
```

```python
# 함수를 별도로 만든 코드
def fix_century(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))


df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)
```

년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
def change_date(x):
    import datetime
    date = x.split(':')[0]
    hour = x.split(':')[1]
    
    if hour == '24':
        hour = '00:00:00'
        # 24시 -> 0시로 표시하면 전날이 되기 때문에 + 1
        finaldate = pd.to_datetime(date+' '+hour) + datetime.timedelta(days=1)
    else:
        hour = hour+':00:00'
        finaldate = pd.to_datetime(date+' '+hour)
    return finaldate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.info()
```

## Merge, Concat

```python
# 국가별 5세이하 사망비율 통계 : https://www.kaggle.com/utkarshxy/who-worldhealth-statistics-2020-complete 데이터 변형

url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv'
df = pd.read_csv(url, index_col='Location')
df.head()
```

91. df1과 df2 데이터를 하나의 데이터 프레임으로 합쳐라

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
pd.concat([df1, df2])
```

92. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해서만 고려한다

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
pd.concat([df3,df4],join='inner')
```

93. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 컬럼을 포함하고, 결측치는 0으로 대체한다

```python
pd.concat([df3, df4], join='outer').fillna(0)
```

94. df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하라

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
pd.merge(df5, df6, how='inner', on='Algeria')
```

95. df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 합집합으로 합쳐라

```python
pd.merge(df5, df6, how='outer', on='Algeria')
```
